```r
dist_Manhattan <- function(p,x,centro){
    dist = 0
    
    dist = (x[1] - centro[1])^p + (x[2] - centro[2])^p
    dist= abs(dist) ^ (1/p)  
            
    return(dist)
}
```

```r
clusters <- function(X,p,centro){
    classe = matrix(0, ncol = 1, nrow = 400)
    
    
    for(i in 1:400){
        distC1 = dist_Manhattan(p, X[i,],centro[1,] )
        distC2 = dist_Manhattan(p, X[i,],centro[2,] )
        
        if(distC1 < distC2){
            classe[i] = 1
        }
        else{
            classe[i] = 2
        }
    }
    return(classe)
    
    
} 
```

```r
updateValueCenter <- function(classe,X){
    row = dim(classe)[1]
    
    XCenterC1 <- 0 
    YCenterC1 <- 0 
    XCenterC2 <- 0 
    YCenterC2 <- 0 
    qtdC1 = 0
    qtdC2 = 0
    
    for(i in 1:row){
        if(classe[i] == 1){
           XCenterC1 = XCenterC1 + X[i,1]
           YCenterC1 = YCenterC1 + X[i,2]
           qtdC1=qtdC1+1
        }
            
        else{
           XCenterC2 = XCenterC1 + X[i,1]
           YCenterC2 = YCenterC2 + X[i,2]
           qtdC2=qtdC2+1
        }
          
    }
    centroC1 = cbind(XCenterC1/qtdC1, YCenterC1/qtdC1)
    centroC2 = cbind(XCenterC2/qtdC2, YCenterC2/qtdC2)
    centro = rbind(centroC1,centroC2)
    return(centro)
    
    
}
```

```r
generateCentro <- function(k){
    centro = matrix(0, ncol = 2, nrow=k)
    for(i in 1:k){
        centro[i,]= rnorm(2, mean=sample(1:5,1), sd=0.3)
    }
    return(centro)
}
clusters <- function(X,p,centro){
    dims = dim(X)[1]
    classe = matrix(0, ncol = 1, nrow = dims)
    qtdCentro <- dim(centro)[1]
    dist = matrix(0, nrow=qtdCentro)
    for(i in 1:dims){
        for(j in 1:qtdCentro){
            dist[j] = dist_Manhattan(p, X[i,],centro[j,] )
            
        }
        p = match(min(dist),dist)
        classe[i] = p

        
    }
    return(classe)
    
} 
```

```r
updateValueCenter <- function(classe,X,k){
    row = dim(classe)[1]
    
    
    centro = matrix(0, nrow = k, ncol=2)
    qtd = matrix(0, nrow=k)
    
    for(i in 1:row){
        for(j in 1:k){
           qtd[j] = sum(classe==j)
           
            if(classe[i] == j){
                centro[j,1] = centro[j,1] + X[i,1]
                centro[j,2] = centro[j,2] + X[i,2]
            }
            
        }
          
    }
    
    for(j in 1:k){
        centro[j,1] =  centro[j,1]/ qtd[j]
        centro[j,2] =  centro[j,2]/ qtd[j]
        
    }
  
    return(centro)
    
    
}

getColors <- function(classe,k){
    color = rainbow(k)
    row = dim(classe)[1]
    colors = matrix('', ncol = row)
    for(i in 1:row){
        for(j in 1:k){
           if(classe[i] == j){
                colors[i] = color[j]
            }
        }
   
    }
    return(colors)
}
```

```r
k = 3
centro = generateCentro(k)
classe = clusters(X,k,centro);
ClassAreDiff = 0
qtd = 0

while(ClassAreDiff == 0 && qtd <100){
   centro = updateValueCenter(classe,X,k)
   classUpdate = clusters(X,k,centro) 
   
   qtdEqual = sum(classUpdate == classe)
   print(qtdEqual)
   qtd = qtd + 1
   
   if(qtdEqual >= (dim(X)[1])){
      ClassAreDiff = 1
      classe = classUpdate  
   }
       
   else 
     classe = classUpdate  

}
```

```r
colors = getColors(classe,3)
plot(X[,1],X[,2],xlim=c(0,6), ylim=c(0,6), col=colors)
par(new=TRUE)
plot(centro[,1], centro[,2],xlim=c(0,6), ylim=c(0,6),pch=7,col='black')
```
